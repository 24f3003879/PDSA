#MASTER THEOREM

Master Theorem is used to find the time complexity of divide and conquer recursive algorithms.

It works for recurrences of the form:

                   T(n)=aT(n/b)+f(n)

                   where a = no of recursive calls
                         b = factor by which input is divided
                         f(n) = extra work

Calculate: n^logb(a) and compare with f(n):

case-1: n^logb(a) > f(n) write n^logb(a)

case-2: n^logb(a) < f(n) write f(n)

case-3: n^logb(a) = f(n) write f(n)*logn



Binary Search

Binary search is an efficient search algorithm used to find the position of a target value within a sorted list. The algorithm compares the target value to the middle element of the list. If the target value is equal to the middle element, the search is complete. Otherwise, the algorithm recursively searches either the left or right half of the list, depending on whether the target value is greater or less than the middle element.

Best Case - O(1)

Worst Case - O(logn)

For binary search, the maximum number of iterations for an array of size $n$ is given by:


$$\lfloor \log_2(n) \rfloor + 1$$for $n = 32$:$$\lfloor \log_2(32) \rfloor + 1 = 5 + 1 = 6$$

```python

def binary_Search(L, v):
  low = 0
  high = len(L) - 1
  while low <= high:
    mid = (low + high) // 2
    if L[mid] < v:
      low = mid + 1
    elif L[mid] > v:
      high = mid - 1
    else:
      return mid
  return False



L = [1, 2, 3, 4, 5, 6, 7, 8, 12, 23, 23, 24, 45, 45, 67, 90]
v = 23
print(binary_Search(L, v))
```

Selection Sort

The main idea:-   select the minimum.

Selection sort is a simple sorting algorithm that works by repeatedly finding the minimum element from an unsorted part of the list and placing it at the beginning of the sorted part of the list.

Best,Average,Worst - O{n^2)

Selection Sort:

Stable? no

In-Place? yes

Insertion Sort

The main idea:- Take one element and insert it where it belongs.

Insertion sort is a simple sorting algorithm that works by repeatedly iterating through the list, comparing adjacent elements and swapping them if they are in the wrong order.

.

Best Case: O(n) if list is already sorted

Average Case and Worst Case: O(n^2)

Stable? yes

In-place? yes

Merge sort is a popular sorting algorithm that uses the divide-and-conquer technique. It works by recursively dividing an list into two halves, sorting each half separately, and then merging them back together into a single sorted list.

Best, Average, Worst Case: O(nlogn)

Stable? yes
